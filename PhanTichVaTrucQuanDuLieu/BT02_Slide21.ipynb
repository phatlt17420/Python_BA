```python
import pandas as pd
import numpy as np
path = 'https://archive.ics.uci.edu/ml/machine-learningdatabases/autos/imports-85.data'
```

```python
df = pd.read_csv(path, names=headers)
```

```python
df['normalized-losses'].replace(to_replace=np.nan, value=mode_[0], inplace=True)
```

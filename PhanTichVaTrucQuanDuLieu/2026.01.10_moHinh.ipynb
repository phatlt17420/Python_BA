```python
import pandas    as pd
from pandas import read_csv

path = 'https://raw.githubusercontent.com/datasethub/ds105/master/Model-Evaluation-and-Refinement.csv'
df = pd.read_csv(path)
```

```python
df
```

```python
x_data = df[['horsepower', 'curb-weight','engine-size','highway-mpg']]
y_data = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42) #chon ngẫu nhiên
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
m1 = LinearRegression()
m1.fit(x_train, y_train)
```

```python
import seaborn as sns
yHat_test = m1.predict(x_test)
ax1 = sns.kdeplot(x=y_test,)
sns.kdeplot(x=yHat_test,ax=ax1)
```

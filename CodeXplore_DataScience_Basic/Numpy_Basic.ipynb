### Array Concatenation and Splitting

```python
x = np.array([1,2,3])
y = np.array([3,2,1])
```

```python
np.concatenate((x,y))
```

```python
grid = np.array([[1,2,3],
                 [4,5,6]])
```

```python
np.concatenate((grid,grid),axis=1) # nối theo cột
```

```python
np.concatenate((grid,grid)) # nối theo dòng
```

```python
x = np.array([1,2,3])
grid = np.array([[9,8,7],
                 [4,5,6]])
np.vstack((x,grid)) # nối khác kích thước theo dòng
```

```python
y = np.array([[2],[2]])
grid = np.array([[9,8,7],
                 [4,5,6]])
np.hstack((y,grid)) # nối khác kích thước theo cột
```

### Linear Algebra (đại số tuyến tính)

```python
np.random.seed(0)
sales_amounts = np.random.randint(0,20, size=(5,3))
```

```python
sales_amounts
```

```python
import pandas as pd
weekly_sales = pd.DataFrame(sales_amounts, index=['Mon','Tue','Wed','Thurs','Fri'], columns = ['Almond Butter',' Peanut Butter', 'Cashew Butter'])
```

```python
weekly_sales
```

```python
# Giá của butter
prices = np.array([10,8,12])
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3), index=['Price'], columns = ['Almond Butter',' Peanut Butter', 'Cashew Butter'])
```

```python
butter_prices
```

```python
total_prices = weekly_sales.dot(butter_prices.transpose())
```

```python
total_prices
```

```python
weekly_sales['Total_prices']=total_prices
```

```python
weekly_sales
```

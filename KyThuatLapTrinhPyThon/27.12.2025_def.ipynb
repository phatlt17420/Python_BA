```python
# Bài 3
import random as r
def initA():
    N = r.randint(30,80)
    return [round(r.uniform(-79,39),2)
            if r.choice([0,1])
            else r.randint(-79,39)
            for _ in range(N) ]

seqA = initA()
print(seqA)
```

```python
def kiem_tra_du_lieu(l):
    for i in l:
        print(type(i))

kiem_tra_du_lieu(seqA)
```

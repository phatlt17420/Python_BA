```python
from IPython.core.display import Math
from tornado.gen import multi


#viết hàm tính giai thừa bằng đệ quy (recursion)
def giai_thua(n):
    if n ==1:
        return 1
    else:
        return n * giai_thua(n-1)
print(giai_thua(5))
```

```python
def tinh_toan (a,b):
    def tinh_tong(a,b):
        return a+b
    return tinh_tong(a,b)
print(tinh_toan(5,6))
```

```python
#Bài 1: Dùng các kỹ thuật tham số dạng *args, **kwargs... thiết kế hàm tính tổng, tích của nhiều số.
import math
def tinhTich (*l):
    p=1
    for i in l:
        p*=i
    return p
print(tinhTich(5,6,1,2))
```

```python
# Bài 2:Viết hàm dùng các kỹ thuật tham số dạng *args, **kwargs... tính chu vi hình vuông, chữ nhật, tròn.
def tinh_chu_vi(ten_hinh, **thong_so) :
    if ten_hinh=='hinh_vuong':
        return thong_so.get('canh') ** 2
    elif ten_hinh == 'hinh_tron' :
        return (thong_so.get('r') **2) * 3.14
    elif ten_hinh == 'hinh_chu_nhat':
        return thong_so.get('a') *thong_so.get('b')
print(tinh_chu_vi('hinh_tron', r =  8))
```

```python
# Bài 3: Viếtt hàm dùng các kỹ thuật tham số dạng *args, **kwargs... tính chu vi hoặc diện tích hình vuông, chữ nhật, tròn.
import math
def tinhToan(**thongSo) :
    if thongSo.get('loaiTinhToan') == 'ChuVi':
        if thongSo.get('loaiHinh') == 'hinh_vuong' :
            return thongSo.get('canh') * 4
        elif thongSo.get('loaiHinh') == 'hinh_tron' :
            return thongSo.get('r') * 2 * math.pi
        elif thongSo.get('loaiHinh') == 'hinh_chu_nhat' :
            return (thongSo.get('a') + thongSo.get('b'))*2
    if thongSo.get('loaiTinhToan') == 'DienTich':
        if thongSo.get('loaiHinh') == 'hinh_vuong' :
            return thongSo.get('canh') **2
        elif thongSo.get('loaiHinh') == 'hinh_tron' :
            return (thongSo.get('r') ** 2) * math.pi
        elif thongSo.get('loaiHinh') == 'hinh_chu_nhat' :
            return thongSo.get('a') * thongSo.get('b')
print(tinhToan(loaiTinhToan='ChuVi',loaiHinh='hinh_tron',r = 2))
```

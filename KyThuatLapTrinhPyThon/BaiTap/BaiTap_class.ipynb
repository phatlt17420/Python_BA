```python
class SinhVien:
    def __init__ (self, mssv,hoten='nguyenVanA', tuoi=20):
        self.mssv = mssv
        self.hoten = hoten
        self.tuoi = tuoi
    def xuat(self):
        print('MSSV:', self.mssv)
        print('Hoten:', self.hoten)
        print('Tuoi:', self.tuoi)
```

```python
svA = SinhVien('12345','Nguyễn Văn A', 20)
svA.xuat()
```

```python
import math as m
class ToaDo:
    def __init__ (self, x,y):
        self.x = x
        self.y = y
    def khoangcach_cach1 (self, B):
        return m.sqrt(pow(self.x, B.x) + pow(self.y, B.y))
    def xuat(self):
        print(f"({self.x:.0f},{self.y:.0f})")
    @staticmethod
    def khoangcach_cach2 (A, B):
        return m.sqrt(pow(A.x, B.x) + pow(A.y, B.y))
    def trungdiem_cach1(self, B):
        return ToaDo((self.x+B.x)/2,(self.y+B.y)/2)
```

```python
A = ToaDo(1,2)
B = ToaDo(3,4)
ToaDo.khoangcach_cach2(A,B)
```

```python
A.trungdiem_cach1(B).xuat()
```

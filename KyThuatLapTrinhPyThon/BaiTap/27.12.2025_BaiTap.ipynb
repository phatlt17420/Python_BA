```python
# Bài 1:
while True:
    try:
        n = int(input("Enter a number: "))
        print(n*n)
        break
    except:
        print('Không phải số nguyên')
        continue
```

```python
# Bài 2:
try:
    a = input("Enter a number: ")
    assert a.find('.') == -1, "Sai Định dạng"
    b = input("Enter b number: ")
    assert a.find('.') == -1, "Sai Định dạng"
    a = float(a)
    b = float(b)
    print(a/b)
except ZeroDivisionError:
    print('Lỗi')
except:
    print('Sai định dạng')

```

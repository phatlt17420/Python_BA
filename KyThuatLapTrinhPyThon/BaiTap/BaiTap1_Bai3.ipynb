```
# Bài tập 3. Cho số tự nhiên A. Hãy tìm số tự nhiên N nhỏ nhất sao cho N lũy thừa N (nhân N cho chính nó N lần) chia hết cho A.
# Hãy viết chương trình tìm số N đó và xuất ra màn hình. Trong đó A có giá trị: 1 ≤ A ≤ 109.
# Ví dụ:
# Nhập: 8 - Xuất: 4.
# Nhập: 13 - Xuất: 13.
```

```
def tich_uoc_nguyen_to(n):
    if n < 2:
        return 0
    tich = 1
    temp = n
    # Kiểm tra số 2
    if temp % 2 == 0:
        tich *= 2
        while temp % 2 == 0:
            temp //= 2
    # Kiểm tra các số lẻ từ 3 đến căn bậc hai của n
    i = 3
    while i * i <= temp:
        if temp % i == 0:
            tich *= i
            while temp % i == 0:
                temp //= i
        i += 2
    # Nếu cuối cùng temp vẫn > 2, thì temp là ước nguyên tố cuối cùng
    if temp > 2:
        tich *= temp
    return tich
```

```

def timSo():
  try:
    A = int(input("Nhập A: "))
  except ValueError:
    print('Nhập sai định dạng')
  z = tich_uoc_nguyen_to(A)
  print(z)
  for i in range(1, z):
    if pow(i,i, A) == 0:
      print(f'Số cần tìm N: {i}')
      break
```

```
def tich_uoc_nguyen_to(n):
    if n < 2: return 0
    tich = 1
    temp = n
    # Xử lý số 2
    if temp % 2 == 0:
        tich *= 2
        while temp % 2 == 0: temp //= 2
    # Xử lý các số lẻ
    i = 3
    while i * i <= temp:
        if temp % i == 0:
            tich *= i
            while temp % i == 0: temp //= i
        i += 2
    if temp > 2: tich *= temp
    return tich

def timSo():
    try:
        A = int(input("Nhập A: "))
        if A <= 0:
            print("Vui lòng nhập số nguyên dương.")
            return
    except ValueError:
        print('Nhập sai định dạng số nguyên.')
        return

    gioi_han = tich_uoc_nguyen_to(A)

    # Chúng ta tìm từ 1 đến gioi_han
    da_tim_thay = False
    for i in range(1, gioi_han + 1):
        # pow(i, i, A) tính (i**i) % A cực nhanh
        if pow(i, i, A) == 0:
            print(f'Số N nhỏ nhất thỏa mãn N^N chia hết cho {A} là: {i}')
            da_tim_thay = True
            break

    if not da_tim_thay:
        print("Không tìm thấy số N thỏa mãn trong phạm vi tính toán.")

timSo()
```

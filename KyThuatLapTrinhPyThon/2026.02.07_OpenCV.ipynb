```python
import cv2
img_gray=cv2.imread('road_image.jpg',cv2.IMREAD_COLOR_BGR)
```

```python
cv2.imshow('Hinh anh xam', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_resize = cv2.resize(img_gray,(300,400))
cv2.imshow('Resize', img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_clip = cv2.getRectSubPix(img_gray,(200,200),(0,0))
cv2.imshow('clip', img_clip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

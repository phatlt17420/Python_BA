```python
#Bài 1
string = 'banana'
from collections import Counter
couter = Counter(string)
print(couter)
```

```python
#Bài 2. Đếm số lần xuất hiện của từ. Cho một danh sách các từ. Hãy in ra 3 từ xuất hiện nhiều nhất.
string2 = ['trần', 'tuấn', 'trần','lê', 'trịnh', 'lê','lê', 'trịnh', 'lê']
couter2 = Counter(string2)
sort_couter2=sorted(couter2.items(),key=lambda item: item[1],reverse=True)
print(sort_couter2[0:3])
```

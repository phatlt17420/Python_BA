```python
from random import random

#tính tiển taxi
distance = int(input("Enter the distance in kilometers: "))
total= 0
if (distance <= 1):
    total += 20
elif (distance <= 3 & distance > 1):
    distance = distance - 1
    total += 20 + distance*13
elif (distance >= 4 & distance < 8):
    total += 13*3 + (distance -3)*12
elif (distance > 8):
    distance -= 8
    total += 13*3 + (4)*12 + distance*10

if total>100:
    total=total-0.08 * total

print (total)
```

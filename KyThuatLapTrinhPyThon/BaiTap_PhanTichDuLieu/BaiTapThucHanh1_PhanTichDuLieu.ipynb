```python
import pandas as pd
import numpy as np
from pygments.lexer import include
```

```python
df = pd.read_csv('StudentPerformance.csv')
```

```python
df.info()
```

## Thống kê các giá trị thiếu

```python
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Tạo DataFrame thống kê
missing_stats = pd.DataFrame({
    'Số lượng thiếu': missing_count,
    'Tỷ lệ (%)': missing_percentage
}).sort_values(by='Số lượng thiếu', ascending=False)

print(missing_stats)
```

```python
df_filter = df.dropna(subset=['Exam_Score']).copy()
```

```python
df_filter.describe(include='all')
```

```python
missing_count = df_filter.isnull().sum()
missing_percentage = (df_filter.isnull().sum() / len(df_filter)) * 100

# Tạo DataFrame thống kê
missing_stats = pd.DataFrame({
    'Số lượng thiếu': missing_count,
    'Tỷ lệ (%)': missing_percentage
}).sort_values(by='Số lượng thiếu', ascending=False)

print(missing_stats)
```

```python
obj_cols = df_filter.select_dtypes(include=['object']).columns

# In ra các giá trị duy nhất để tìm ký tự lạ
for col in obj_cols:
    print(f"Cột {col}: {df_filter[col].unique()}") # Xem 5 giá trị đầu tiên của mỗi cột
```

```python
## thực hiện xử lý các dấu ?? và chuẩn hoá các dữ liệu
df_filter[obj_cols] = df_filter[obj_cols].replace(r'\?\?', '', regex=True)
 ##Xóa khoảng trắng thừa phát sinh sau khi xóa dấu ??
for col in obj_cols:
    df_filter[col] = df_filter[col].astype(str).str.strip()
```

```python
df_filter[obj_cols] = df_filter[obj_cols].replace(['nan', 'NaN'], np.nan)
```

```python
missing_count = df_filter.isnull().sum()
missing_percentage = (df_filter.isnull().sum() / len(df_filter)) * 100

# Tạo DataFrame thống kê
missing_stats = pd.DataFrame({
    'Số lượng thiếu': missing_count,
    'Tỷ lệ (%)': missing_percentage
}).sort_values(by='Số lượng thiếu', ascending=False)

print(missing_stats)
```

```python
df_filter.describe(include='all') ## Bộ dự liệu đã được làm sạch (xoá các ký tự đặc biệt, chuẩn hoá các giá trị obj, các giá trị trống chuyển thành NaN)
```

```python
# Qua thống kê nhận thấy bộ dữ liệu có các giá trị < 0 vô lý. THực hiện thống kê lại các giá trị số vô lý.
total_rows = len(df_filter)
num_cols = df.select_dtypes(include=['number']).columns

negative_stats = []
for col in num_cols:
    neg_count = (df_filter[col] < 0).sum()
    if neg_count > 0:
        negative_stats.append({
            'Cột': col,
            'Số lượng âm': neg_count,
            'Tỷ lệ (%)': (neg_count / total_rows) * 100
        })

print(pd.DataFrame(negative_stats))
```

```python
# 3. Sử dụng hàm .clip(lower=0)
# Lệnh này sẽ: Giữ nguyên nếu >= 0, và chuyển thành 0 nếu < 0
cols_to_fix = ['Hours_Studied', 'Tutoring_Sessions', 'Physical_Activity']
df_filter[cols_to_fix] = df_filter[cols_to_fix].clip(lower=0)
```

```python
df_filter.describe(include='all')
```

```python
# Bộ dữ liệu cơ bản đã được xử lý => thực hiện thống kê và điến khuyết các giá trị null
missing_count = df_filter.isnull().sum()
missing_percentage = (df_filter.isnull().sum() / len(df_filter)) * 100

# Tạo DataFrame thống kê
missing_stats = pd.DataFrame({
    'Số lượng thiếu': missing_count,
    'Tỷ lệ (%)': missing_percentage
}).sort_values(by='Số lượng thiếu', ascending=False)

print(missing_stats)
```

```python
## Với tỷ lệ khuyết thấp thỉ sử dụng Trung vị (Median) cho biến số và Mode cho biến chữ.
#Diền khuyết cho cột số
for col in num_cols:
    median_val = df_filter[col].median()
    df_filter[col] = df_filter[col].fillna(median_val)
#Diền khuyết cho cột obj
for col in obj_cols:
    mode_val = df_filter[col].mode()[0] # Lấy giá trị xuất hiện nhiều nhất
    df_filter[col] = df_filter[col].fillna(mode_val)
```

```python
# Bộ dữ liệu cơ bản đã được xử lý => thực hiện thống kê và điến khuyết các giá trị null
missing_count = df_filter.isnull().sum()
missing_percentage = (df_filter.isnull().sum() / len(df_filter)) * 100

# Tạo DataFrame thống kê
missing_stats = pd.DataFrame({
    'Số lượng thiếu': missing_count,
    'Tỷ lệ (%)': missing_percentage
}).sort_values(by='Số lượng thiếu', ascending=False)

print(missing_stats)
```

## Bước 3: Phân tích khám phá bộ dữ liệu (EDA)

### 3.1: Thông kê mô tả từng biến

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=len(num_cols), ncols=2, figsize=(15, 20))
for i, var in enumerate(num_cols):
    # 1. Vẽ Histogram + KDE
    sns.histplot(df_filter[var], kde=True, color='teal', bins=30, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram & KDE của {var}')

    # 2. Vẽ Boxplot ở cột bên phải
    sns.boxplot(x=df_filter[var], color='orange', ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot của {var}')

plt.tight_layout()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Thiết lập lưới biểu đồ
fig, axes = plt.subplots(nrows=len(obj_cols), ncols=2, figsize=(15, 45))
sns.set_theme(style="whitegrid")

total = len(df_filter) # Tổng số dòng để tính tỷ lệ

for i, var in enumerate(obj_cols):
    # 1. Countplot: Vẽ biểu đồ cột
    ax1 = axes[i, 0]
    sns.countplot(data=df_filter, x=var, ax=ax1, hue=var, palette='viridis', legend=False)
    ax1.set_title(f'Phân bổ số lượng & Tỷ lệ theo {var}')
    ax1.set_xlabel('')

    # --- THÊM TỶ LỆ PHẦN TRĂM ---
    for p in ax1.patches:
        percentage = f'{100 * p.get_height() / total:.1f}%' # Tính %
        x = p.get_x() + p.get_width() / 2 # Vị trí x giữa cột
        y = p.get_height() + (total * 0.01) # Vị trí y trên đỉnh cột một chút
        ax1.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10, fontweight='bold')

    # 2. Boxplot: So sánh với điểm thi
    ax2 = axes[i, 1]
    sns.boxplot(data=df_filter, x=var, y='Exam_Score', ax=ax2, hue=var, palette='Set2', legend=False)
    ax2.set_title(f'Biến thiên Điểm thi theo {var}')
    ax2.set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
# 1. Tính toán ma trận tương quan
corr_matrix = df_filter.select_dtypes(include=['float64', 'int64']).corr()

# 2. Thiết lập khung hình
plt.figure(figsize=(10, 8))

# 3. Vẽ Heatmap với tông màu sáng và nhẹ nhàng hơn
sns.heatmap(corr_matrix,
            annot=True,
            fmt=".2f",
            cmap='YlGnBu',     # Bảng màu sáng (Vàng - Xanh)
            linewidths=1,      # Tăng độ rộng đường kẻ giữa các ô
            linecolor='white', # Đường kẻ màu trắng cho sạch sẽ
            cbar_kws={"shrink": .8}) # Thu nhỏ thanh chú thích màu một chút

plt.title('Ma trận tương quan giữa các biến số', fontsize=15, pad=20)
plt.show()
```

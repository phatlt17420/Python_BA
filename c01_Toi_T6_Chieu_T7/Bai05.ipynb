```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
data = df['price'][(df['price']<Q1-1.5*IQR)|(df['price']>Q3+1.5*IQR)]
print(data)
```

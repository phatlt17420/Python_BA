## Hàm plot trong dataFrame

```python
data = {
    'a': [1,2,3,4,5,6],
    'b': [3,4,5,8,12,12],
    'c': [35,45,55,82,12,24],
}
df1 = pd.DataFrame(data)
```

```python
df1.plot(kind='line')
```

```python
# chuyển 1 cột thành index
df1.set_index('a', inplace=True)
```

```python
df1.plot(kind='line')
```

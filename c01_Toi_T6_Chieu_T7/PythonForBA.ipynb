## Hàm

```python
def tinh(a, b):
    cong = a + b
    tru = a - b
    nhan = a * b
    chia = a / b
    if (a > b):
        return cong, tru
    else:
        return nhan, chia
```

```python
x, y = tinh(3, 4)
```

```python
x
```

```python
y
```

```python
def tinh(a: int, b: float) -> tuple:  # gợi ý kiểu dữ liệu đưa vào, và trả ra
    cong = a + b
    tru = a - b
    nhan = a * b
    chia = a / b
    if (a > b):
        return cong, tru
    else:
        return nhan, chia
```

```python
tinh(5.5, 2.5)
```

```python
def func(x, y=1, *args, **kwargs):  # x bắt buộc truyền vào, y không bắt buộc truyền, default =1
    ...
    ...
    return x, y, args, kwargs
```

```python
def tinh(a=2, b=1):
    cong = a + b
    tru = a - b
    nhan = a * b
    chia = a / b
    return cong, tru, nhan, chia
```

```python
tinh()
```
